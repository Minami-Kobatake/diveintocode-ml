**【問題1】べき乗の算術演算子を使用した実装**


```python
# 紙を43回折った時の厚さを計算するコード

THICKNESS = 0.00008
folded_thickness = THICKNESS * 2 **43
print("厚さ： {}メートル".format(folded_thickness))
```

**【問題2】単位の変換**

```python
# メートルを万キロメートルに変換して小数点以下2桁で表示する
print("厚さ： {:.2f}万キロメートル".format(folded_thickness / 10000000))
```

**【問題3】for文を使用して作成**

```python
THICKNESS = 0.00008
folded_thickness = THICKNESS * 2

for i in range(1, 44, 1):  # 1~43の範囲

  if i < 43:
    folded_thickness *= 2
  else:
    print("紙を {} 回折った時の厚みは {:.2f} 万キロメートルです".format(i, folded_thickness))
```

**【問題4】計算時間の比較**

```python
# べき乗Ver.
import time
start = time.time()

THICKNESS = 0.00008
folded_thickness = THICKNESS * 2 **43
print("厚さ： {:.2f}万キロメートル".format(folded_thickness / 10000000))

#####
elapsed_time = time.time() - start
print("time : {}[s]".format(elapsed_time))
```

```python
# for文Ver.
import time
start = time.time()
#####

THICKNESS = 0.00008
folded_thickness = THICKNESS * 2

for i in range(1, 44, 1):  # 1~43の範囲

  if i < 43:
    folded_thickness *= 2
  else:
    print("厚さ： {:.2f}万キロメートル".format(folded_thickness / 10000000))

#####
elapsed_time = time.time() - start
print("time : {}[s]".format(elapsed_time))
```

**【問題5】リストへの保存**

```python
folded_thickness_list =[]

THICKNESS = 0.00008
folded_thickness = THICKNESS

for i in range(0, 44, 1):  # 1~43の範囲
  folded_thickness_list . append(folded_thickness)
  print("紙を {} 回折った時の厚みは {} 万キロメートルです".format(i, folded_thickness))
  
  folded_thickness = folded_thickness*2

print(len(folded_thickness_list))
print(folded_thickness_list)
```

**【問題6】折れ線グラフの表示**

```python
import matplotlib.pyplot as plt
%matplotlib inline
# グラフを表示する。タイトルと軸ラベル名付き。
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness [m]")
plt.plot(folded_thickness_list)  # 「リスト名」のところにリストの変数名を入れる
plt.show()
```

**【問題7】グラフのカスタマイズ**

```python
plt.title("thickness of folded paper", fontsize=15)
plt.xlabel("number of folds", fontsize=12)
plt.ylabel("thickness[10000km]", fontsize=12)
plt.tick_params(labelsize=15) # 軸の値に関する設定を行う
plt.plot(folded_thickness_list, lw=5, label="0.00008")
plt.show()
```
