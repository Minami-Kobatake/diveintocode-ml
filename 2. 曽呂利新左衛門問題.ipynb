**【問題1】100日目の米粒の数**

100日目には合計で米粒をいくつもらっているのかを計算するコードを作成し、答えを出力してください。

また、1日目から100日目までの「その日にもらう米の数」と「その日までに累計で貰う米の数」の推移を折れ線グラフで表現してください。（縦軸が米の数、横軸が日数）

```python
'''
1日目
    x = 1           累計(2**0 = 1)
2日目
    x = 1**2 = 2  累計(2**0 + 2**1 =3)
3日目
    x = 2**2 = 4  累計(2**0 + 2**1 + 2**2 = 7)
4日目
    x = 2**3 = 8  累計(2**0 + 2**1 + 2**2 + 2**3 = 15)
n日目
    x = 2**(n-1)  累計(2**0 +2**1 + 2**2 ・・・ 2**(n-1)= )
'''

#100日目の米粒の数
number_of_rice_grains = 2 ** 99
print(number_of_rice_grains)
```

```python
LAST_DAY = 100
number_of_rice_grains = 2 **(LAST_DAY-1)

print(number_of_rice_grains)
```

```python
# 1日目から100日目までの「その日にもらう米の数」

for days in range(1, 101, 1): #1~100の数
  number_of_rice_grains = 2 **(days-1)
  print("{}日目の米粒の数は{}粒です。".format( days, number_of_rice_grains))
  days += 1
```

```python
# 1日目から100日目までの「その日までに累計で貰う米の数」
days = 1
cumulant_amount = 2 **(days-1)
print("{}日目までの米粒の累積数は{}粒です。".format( days, cumulant_amount))

for days in range(1, 100, 1): #1~99の数
  days += 1
  cumulant_amount += 2**(days-1)
  print("{}日目までの米粒の累積数は{}粒です。".format( days, cumulant_amount))
  
  
```

```python
# 1日目から100日目までの「その日にもらう米の数」をリストに入れる

number_of_rice_grains_list =[]

for days in range(1, 101, 1): #1~100の数
  
  number_of_rice_grains = 2 **(days-1)
  number_of_rice_grains_list.append(number_of_rice_grains)
  days += 1

print(number_of_rice_grains_list)
print(len(number_of_rice_grains_list))
```

```python
# 1日目から100日目までの「その日にもらう米の数」をグラフに表す

import matplotlib.pyplot as plt
%matplotlib inline
# グラフを表示する。タイトルと軸ラベル名付き。
plt.title("number of rice grants")
plt.xlabel("days")
plt.ylabel("number of rice grants")
plt.plot(number_of_rice_grains_list)  # 「リスト名」のところにリストの変数名を入れる
plt.show()
```

```python
# 1日目から100日目までの「その日までに累計で貰う米の数」をリストに入れる

cumulant_amount_list = []
days = 1
cumulant_amount = 2 **(days-1)

cumulant_amount_list.append(cumulant_amount)

for days in range(1, 100, 1): #1~99の数
  days += 1
  cumulant_amount += 2**(days-1)
  cumulant_amount_list.append(cumulant_amount)
  

print(cumulant_amount_list)
print(len(cumulant_amount_list))
```

```python
# 1日目から100日目までの「その日までに累計で貰う米の数」をグラフに表す

import matplotlib.pyplot as plt
%matplotlib inline
# グラフを表示する。タイトルと軸ラベル名付き。
plt.title("Cumulative number of rice grains")
plt.xlabel("days")
plt.ylabel("Cumulative number of rice grains")
plt.plot(cumulant_amount_list)  # 「リスト名」のところにリストの変数名を入れる
plt.show()
```
